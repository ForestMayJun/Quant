```python
import numpy as np
import pandas as pd

def conti_up_down(price: pd.DataFrame, T=30, is_up=True):
    A = np.triu(np.ones((T, T)))

    def _conti_up_down(price_daily, T=T, is_up=is_up):
        return_daily = price_daily.pct_change().fillna(0)

        def _for_stock(r_daily_stock):
            conv_results = np.apply_along_axis(
                lambda x: np.convolve(x, A, mode='valid'), axis=0, arr=r_daily_stock
            )

            if is_up:
                res = np.max(conv_results, axis=1)
            else:
                res = np.min(conv_results, axis=1)
            
            return pd.Series(np.concatenate(([np.nan] * (T - 1), res)), index=r_daily_stock.index)
        
        return return_daily.apply(_for_stock)
    
    return price.groupby(level='Date').apply(_conti_up_down)
```

## 结合成交量数据的单日涨跌度

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
tqdm.pandas()
# pd.set_option('display.max_rows', 50)
```

```python
vol = pd.read_hdf('/mnt/datadisk2/aglv/aglv/lab_aglv/vol_2021.h5')
open2021 = pd.read_hdf('/mnt/datadisk2/aglv/aglv/lab_aglv/open_2021.h5')
# vol = vol.set_index(['Date', 'EndTime'])
open2021 = open2021.set_index(['Date', 'EndTime'])
```

```python
vol
```

```python
open2021
```

```python
def _to_percent(vol_daily):
    return vol_daily.apply(lambda x : x / x.sum())
vol_percent = vol.groupby(level='Date').progress_apply(_to_percent)
```

```python
vol_percent
```

```python
vol_stack = vol_percent.stack()
```

```python
vol_stack.index.names = ['Date', 'EndTime', 'Stock']
```

```python
vol_stack
```

```python
open_stack = open2021.stack()
open_stack.index.names = ['Date', 'EndTime', 'Stock']
open_stack
```

```python
open_vol = pd.concat([vol_stack, open_stack], axis=1)
```

## 日内极端收益前后的反转特性与因子构建
$ S = |x - median(x)|$

```python
def extre_return(price:pd.DataFrame):
    def _extre_return(price_daily:pd.DataFrame):
        daily_return = price_daily.pct_change().fillna(0)
        return (daily_return - daily_return.median()).abs().max()
    
    return price.groupby(level='Date').progress_apply(_extre_return)
```

```python
def extre_return_plus(price:pd.DataFrame, p=1):
    def _extre_return_plus(price_daily:pd.DataFrame):
        daily_return = price_daily.pct_change().fillna(0)
        daily_return.index = range(len(daily_return))
        daily_max_id = (daily_return - daily_return.median()).idxmax()
        # id = (daily_max_id - p).where(daily_max_id - p > -1).fillna(0).astype('Int64')
        def assist(stock):
            stock_id = daily_max_id[stock.name]
            return stock.iloc[stock_id - p: stock_id + p].sum() if stock_id > p else stock.iloc[0]

        return daily_return.apply(assist)
    
    return price.groupby(level='Date').progress_apply(_extre_return_plus)
```

### 处理分钟级数据 

按某个价格提取成为一个以股票代码成列 时间成行的二维df

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt

# root = "/mnt/datadisk/pjluo_shared/minute_h5_data/20171102.h5"
# df = pd.read_hdf(root)
# df
```

```python
import os
import re
from pathlib import Path

folder_root = '/mnt/datadisk/pjluo_shared/minute_h5_data/'
folder = Path(folder_root)
f_names = [f.name for f in folder.iterdir() if f.is_file() and f.name.endswith('.h5')]
f_2022 = [f for f in f_names if f.startswith('2022') and len(f) == 11]
f_2022.sort(key=lambda x: int(x[:8]))
f_2022
```

```python
from tqdm.notebook import tqdm
df_dic = {}
for f in tqdm(f_2022):
    df = pd.read_hdf(folder_root + f)
    df_reset = df.reset_index()
    df_multi = df_reset.set_index(['InstrumentID', 'EndTime'])
    df_dic[f[:8]] = df_multi
```

```python
df_dic
```

```python
p_2022 = pd.concat(df_dic.values(), keys=df_dic.keys())
p_2022
```

```python
p_2022.index.names = ['Date','InstrumentID', 'EndTime']
```

```python
p_2022
```

```python
open2022 = p_2022['Open']
open2022
```

```python
p_2022_unstack = open2022.unstack(level='InstrumentID')
p_2022_unstack
```

```python
p_2022_unstack.to_hdf('/mnt/datadisk2/aglv/aglv/lab_aglv/close_2022.h5', key='df')
```

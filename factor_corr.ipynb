```python
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde
from scipy.special import roots_chebyt

series = close_pct.iloc[1][:30]
f = gaussian_kde(series)
df = pd.DataFrame()
df['x'] = pd.Series(np.arange(-0.15, 0.15, 0.01))
df['y'] = f(df['x'])
def g(f, a, b):
    def lambda_func(x):
        return (b-a) / 2 * f((b-a)*x/2 + (b+a)/2)
    
    return lambda_func
    
nodes, weights = roots_chebyt(len(df['x']))
df['F'] = np.array([np.sum(np.dot(weights, g(f, -5, sample)(nodes))) for sample in df['x']])
df['f2'] = f(series)

df
```

```python
import numpy as np
pd.set_option('display.max_rows', 150)
from tqdm.notebook import tqdm
tqdm.pandas()

def vwap_corr_stock(df:pd.Series, min_gap=10, type='Close'):
    # tl = time_list_t0(min_gap=min_gap)
    # n_tl = len(tl)
    df = df.sort_values('EndTime')
    n_time = int(len(df)/min_gap)
    v_wap = []
    for i in range(n_time):
        v_wap.append(
            np.dot(df[type][min_gap*i:min_gap*(i+1)], df['LastVolume'][min_gap*i:min_gap*(i+1)]) / np.sum(df['LastVolume'][min_gap*i:min_gap*(i+1)]) if np.sum(df['LastVolume'][min_gap*i:min_gap*(i+1)]) != 0 else np.nan
        )
    v_wap = pd.Series(v_wap)
    v_wap = (v_wap - v_wap.mean()) / (v_wap.std() + 1e-10)

    return v_wap.corr(v_wap.shift(-1))

def vwap_corr_df(df:pd.DataFrame):
    s_list = pd.Series(df.index.unique(), index=df.index.unique())
    return s_list.progress_apply(lambda x: vwap_corr_stock(df.loc[x]))

root = '/mnt/datadisk/pjluo_shared/minute_h5_data/20230724_new.h5'
df = pd.read_hdf(root)
# pd.Series(df.index.unique(), index=df.index.unique())
dfs = vwap_corr_df(df)
dfs
```

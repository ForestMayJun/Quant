```python
import pandas as pd
import numpy as np

root = '/mnt/datadisk2/aglv/aglv/lab_aglv/aglv_factor/s_phi_v2_230101_240101_r30.csv'
# print(df)
df = pd.read_csv(root)
df.replace(np.inf, 0, inplace=True)
df.replace(-np.inf, 0, inplace=True)
# df.replace(0, np.nan, inplace=True)

df = df.iloc[::-1].reset_index(drop=True)
print(df)
# df = pd.read_csv(root)
# df.replace(np.inf, 0, inplace=True)
# df.replace(-np.inf, 0, inplace=True)
# print(df)
print(df.mean(), df.std())
```

```python
from tqdm.notebook import tqdm
tqdm.pandas()
def zscore(series, window=30):
    # zs_series = pd.Series([(series[i] - series.iloc[:i+1].mean()) / (series.iloc[:i+1].std() + 1e-10) for i in range(len(series))])
    for i in range(window, len(series)):
        series.iloc[i] = (series.iloc[i] - series.iloc[i-window:i].mean()) / (series.iloc[i-window:i].std() + 1e-10)
    return series

# s = pd.Series([1, 3, 5, 1, 2])
# print(s[3:5])
# print(zscore(s))

data = df.iloc[:, 2:4].copy().progress_apply(zscore)
data['date'] = df['date']

data
```

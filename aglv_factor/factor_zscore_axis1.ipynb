对某一时刻截面中所有股票因子值作为样本空间进行zscore

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

root = '/mnt/datadisk2/aglv/aglv/lab_aglv/aglv_factor/s_phi_v2_230101_240101_r30.csv'
df = pd.read_csv(root)
df.replace(np.inf, 0, inplace=True)
df.replace(-np.inf, 0, inplace=True)
df = df.iloc[::-1].reset_index(drop=True)
df
```

```python
sub_df = df.iloc[:, 1:].apply(zscore, axis=1)
sub_df = pd.DataFrame([pd.Series(array) for array in sub_df])
sub_df
```

```python
df.iloc[:, 1:] = sub_df.values
df = df.fillna(0)
df
```

```python
print(df.iloc[:,1:].mean(axis=1))
print(df.iloc[:,1:].std(axis=1))
```

```python
df.iloc[:, 1:].std(axis=0)
```

```python
df.to_csv('s_phi_v2_zs_230101_240101_r30.csv', index=False)
```
